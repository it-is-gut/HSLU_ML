## Install Python

* Download and install [Python](https://www.python.org/downloads/) version 3

* Alternatively you can use [Anaconda](https://www.anaconda.com/)

When installing python, make sure that you install **pip** and **add python to your environment variables**:


### Install Packages

For the exercises the following packages are required:

You can install or upgrade them with the following command using pip:
* `pip3 install --upgrade jupyter numpy pandas pandas-profiling scikit-learn matplotlib seaborn mlxtend nltk scikit-image tqdm plotly scipy jupyter_contrib_nbextensions`

### Enable Jupyter Extensions

Next run the following command to install the jupyter extensions:

* `jupyter contrib nbextension install --user`

### Manage extensions within Jupyter Notebook
Instead of using the command line you can use the jupyter notebook directly to manage plugins.

* Launch a jupyter notebook and open the tab *Nbextensions*
* Make sure that the *exercise2* plugin is enabled

### Most important shortcuts

* `Esc`: go to command mode
* `Enter`: go to edit mode
* `m`: switch cell to markdown
* `y`: switch cell to code
* `Shift-Enter`: Execute cell
* `Alt-Enter`: Execute cell and create new
* `dd`: delete cell
* `a`: Insert new cell above
* `b`: Insert new cell below
* `k`: Move to cell above
* `j`: Move to cell below
